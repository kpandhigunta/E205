# Visualizing Points in main.cpp

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([
    [0,0],
    [20,20],
    [40,18],
    [41,10],
    [55,25],
    [80,15],
    [65,-5],
    [45,-8],
    [38,-20],
    [28,-18],
    [0,0]
]).T
o1 = np.array([
    [50,15],
    [56,20],
    [60,15],
    [56,10],
    [50,15],
]).T
o2 = np.array([
    [54,5],
    [66,5],
    [62,-2],
    [56,0],
    [54,5]
]).T
```

```python
bluehex = '#1f77b4'
orangehex = '#ff7f0e'
plt.plot(points[0], points[1], color=bluehex)
plt.plot(o1[0], o1[1], color=orangehex)
plt.plot(o2[0], o2[1], color=orangehex)
plt.xlabel('x (meters)')
plt.ylabel('y (meters)')

plt.show()
```
